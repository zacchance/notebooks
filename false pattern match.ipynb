```python
def plotEllipse(pos,P,edge,face): 
    U, s , Vh = np.linalg.svd(P) 
    orient = math.atan2(U[1,0],U[0,0]) 
    ellipsePlot = pa.Ellipse(xy=pos, width=math.sqrt(s[0]), 
    height=math.sqrt(s[1]), angle=orient,facecolor=face, linestyle='dashed', edgecolor=edge) 
    ax = pl.gca() 
    ax.add_patch(ellipsePlot); 
    return ellipsePlot 
```

# Probability of False Pattern Match

## Introduction

Consider a given set of random points $\bx_i \in \mathbb{R}^2$ for $i = 1,2,\ldots,N$. Each $\bx_i$ is Gaussian distributed
such that $\bx_i \sim \mathcal{N}(\bm_i,\bP)$ where the covariance matrix $\mathbf{P}$ the mean vectors $\bm_i$ are known.  Let the total set of given points be defined as

$$\cX = \{\bx_i: i = 1,2,\ldots,N\}.$$

The points are chosen independently such that their joint probability density function can be written as

$$p(\bx_1,\bx_2,\ldots,\bx_N) = \prod_{i=1}^N p(\bx_i).$$




### Example: Given pattern

```python
N = 5
sigma_x = .5
mean_vec = np.array([[-1, 1],[0, 0],[1, -1],[1, 1],[-1,-1]])
#x_i = np.random.randn(N,2)
x_i = mean_vec
P = sigma_x**2 * np.eye(2)
```

```python
max_x = np.ndarray.max(abs(x_i))
edge = max_x + 2.0*sigma_x

pl.figure(figsize = (7,7))

for n in range(N):
    plotEllipse([x_i[n,0],x_i[n,1]],P,'b','none')
    pl.scatter(x_i[n,0],x_i[n,1])
    pl.text(x_i[n,0]+sigma_x/2,x_i[n,1],'$\mathbf{x}_' + str(n + 1) + '$',fontsize = 17)
    
pl.plot([-edge,edge], [0, 0],'k--')
pl.plot([0, 0],[-edge,edge],'k--')
pl.title('given pattern',fontsize = 15)
pl.ylabel('$x_1$',fontsize = 20)
pl.xlabel('$x_2$',fontsize = 20)
pl.axis('scaled')
pl.axis([-edge, edge, -edge, edge]);
```

## Nuisance Points

Now assume that the $N$ points in the given pattern have been embedded in a set of $M$ random auxillary points $\by_j$ for $j = 1,2,\ldots,M$; denote the entire set of auxillary points as 

$$\begin{equation}
\cY = \{\by_j: j = 1,2,\ldots,M\}.
\end{equation}$$ 

The auxillary points $\by_j$ are also independently drawn and Gaussian distributed such that $\by_j \sim \cN(\bu_j,\bQ)$ where the covariance matrix $\bQ$ is known and the mean vectors $\bu_j$ are random with known density $p(\bu_j)$.

### Example: Total scene

```python
M = 50
sigma_y = .2
sigma_u = 2
y_j = sigma_u**2 * np.random.randn(M,2)
Q = sigma_y**2 * np.eye(2)
```

```python
max_y = np.ndarray.max(abs(y_j))
total_max = max(max_x,max_y)
sigma_max = max(sigma_x,sigma_y)
edge = total_max + 2.0*sigma_max

pl.figure(figsize = (10,10))
for n in range(N):
    plotEllipse([x_i[n,0],x_i[n,1]],P,'b','none')
    pl.scatter(x_i[n,0],x_i[n,1],c='b')
    pl.text(x_i[n,0]+sigma_x/2,x_i[n,1],'$\mathbf{x}_{' + str(n + 1) + '}$',fontsize = 17)

for m in range(M):
    plotEllipse([y_j[m,0],y_j[m,1]],Q,'r','none')
    pl.scatter(y_j[m,0],y_j[m,1],c='r')
    pl.text(y_j[m,0]+sigma_y/2,y_j[m,1],'$\mathbf{y}_{' + str(m + 1) + '}$',fontsize = 17)    
    
pl.plot([-edge,edge], [0, 0],'k--')
pl.plot([0, 0],[-edge,edge],'k--')
pl.title('pattern and nuisance points',fontsize = 15)
pl.ylabel('$x_1$',fontsize = 20)
pl.xlabel('$x_2$',fontsize = 20)
pl.axis('scaled')
pl.axis([-edge, edge, -edge, edge]);
```

## Pattern Matching

The goal of pattern matching is to find the subset of all points $\cX \cup \cY$ that correspond to the original given pattern $\cX$.  The _observed points_ $\bz_\ell$ for $\ell = 1,2,\ldots,M+N$ are defined as an unknown permutation of the total set of points $\cX \cup \cY$ such that

$$\bz_\ell = \pi([\bx, \by]).$$
